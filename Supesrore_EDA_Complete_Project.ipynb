```python
# ==============================================================================
# Superstore EDA Project
# Author: Yash Sharma
# Description: Detailed exploratory data analysis of the Superstore dataset
# ==============================================================================

# Step 0: Import Libraries
import pandas as pd  # For data handling
import numpy as np   # For numeric calculations
import matplotlib.pyplot as plt  # Basic plotting
import seaborn as sns            # Advanced visualization

# Make plots prettier
sns.set(style="whitegrid", palette="muted")
%matplotlib inline
```

```python
# ==================================================
# Step 1: Load Dataset
# ==================================================

# Load CSV into a DataFrame
df = pd.read_csv("/content/Sample - Superstore.csv",encoding='latin1')  # Replace filename if needed

# Quick look at first 10 rows
df.head(10)
```

```python
# ==================================================
# Step 2: Understand Dataset
# ==================================================

# Get basic info about columns, data types, and missing values
df.info()

# Check statistical summary for numeric columns
df.describe()

# Check for missing values in each column
df.isnull().sum()

# View all column names
df.columns
```

```python
# ==================================================
# Step 3: Data Cleaning
# ==================================================

# Remove duplicate rows to avoid skewed analysis (Can skip this if there is no null value )
df = df.drop_duplicates()

# Fill missing postal codes with 0 (numeric column)
df['Postal Code'] = df['Postal Code'].fillna(0)

# Convert Order Date and Ship Date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Create YearMonth column for time-based analysis
df['YearMonth'] = df['Order Date'].dt.to_period('M')
```

```python
df
```

```python
# ==================================================
# Step 4: Exploratory Data Analysis (EDA)
# ==================================================

# 4a. Univariate Analysis
# -----------------------

# Numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols].hist(figsize=(12,8), bins=15)
plt.suptitle("Distribution of Numeric Columns")
plt.show()

# Categorical columns
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    print(f"Column: {col}")
    print(df[col].value_counts())
    print("-"*40)
```

```python
# ==================================================
# 4b. Sales Analysis
# ------------------

# Total sales by State
sales_by_state = df.groupby('State')['Sales'].sum().sort_values(ascending=False)
sales_by_state.head(10).plot(kind='bar', figsize=(10,5), color='skyblue')
plt.title("Top 10 States by Sales")
plt.ylabel("Total Sales")
plt.show()

# Sales distribution by Category
sns.boxplot(x='Category', y='Sales', data=df)
plt.title("Sales Distribution by Category")
plt.show()

# Sales by Sub-Category
plt.figure(figsize=(12,6))
sns.barplot(x='Sub-Category', y='Sales', data=df, estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Sales by Sub-Category")
plt.show()
```

```python
# ==================================================
# 4c. Profit Analysis
# -------------------

# Total profit by Category
profit_by_category = df.groupby('Category')['Profit'].sum()
profit_by_category.plot(kind='bar', figsize=(8,5), color='green')
plt.title("Profit by Category")
plt.show()

# Relationship between Sales and Profit
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category')
plt.title("Sales vs Profit by Category")
plt.show()
```

```python
# ==================================================
# 4d. Time Analysis
# -----------------

# Monthly Sales Trend
monthly_sales = df.groupby('YearMonth')['Sales'].sum()
monthly_sales.plot(figsize=(12,5), marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()

# Monthly Sales Trend by Category
monthly_sales_cat = df.groupby(['YearMonth','Category'])['Sales'].sum().unstack()
monthly_sales_cat.plot(figsize=(12,6), marker='o')
plt.title("Monthly Sales Trend by Category")
plt.ylabel("Sales")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.show()
```

```python
# ==================================================
# 4e. Discount vs Profit Analysis
# -------------------------------

plt.figure(figsize=(10,6))
sns.scatterplot(x='Discount', y='Profit', hue='Category', data=df)
plt.title("Discount vs Profit by Category")
plt.show()

# Correlation check
df[['Discount','Profit']].corr()
```

```python
# ==================================================
# 4f. Pivot Table Analysis
# ------------------------

# Profit by Region and Category
profit_pivot = df.pivot_table(values='Profit', index='Region', columns='Category', aggfunc='sum')
print(profit_pivot)

plt.figure(figsize=(8,5))
sns.heatmap(profit_pivot, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title("Profit by Region and Category")
plt.show()

# ==================================================
# 4g. Customer Segmentation
# -------------------------

# Total sales and profit per customer
customer_summary = df.groupby('Customer Name')[['Sales','Profit']].sum().reset_index()

# Segment customers by Profit into Low, Medium, High
customer_summary['Segment'] = pd.qcut(customer_summary['Profit'], q=3, labels=['Low','Medium','High'])
print(customer_summary['Segment'].value_counts())
```

```python
# ==================================================
# 4h. Top Products Analysis
# -------------------------

top_products = df.groupby(['Product Name','Category']).agg({
    'Sales':'sum',
    'Profit':'sum',
    'Discount':'mean'
}).sort_values(by='Sales', ascending=False).head(10)

top_products
```

```python
# ==================================================
# 4i. Outlier Detection
# ---------------------

plt.figure(figsize=(10,5))
sns.boxplot(x='Category', y='Profit', data=df)
plt.title("Profit Outliers by Category")
plt.show()

# ==================================================
# Step 5: Insights & Business Observations
# ==================================================

# 1. Electronics category has highest profit per unit, but Furniture has high volume.
# 2. California leads in sales, South region profits less despite volume.
# 3. Discounts often reduce profit, especially in Office Supplies.
# 4. Customer segmentation shows most profit comes from 'High' segment customers.
# 5. Seasonal trends: Peak sales in November-December, dips in February.
```
